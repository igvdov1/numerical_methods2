```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(t):
    return np.sin(5 * t) +  0.002 * np.cos(25.5 * t)
```

```python
h = 1/(2*(1+0.002))
grid = np.arange(0, 2*np.pi, h)
res = np.fft.fft(f(grid))

plt.plot(grid, f(grid))
plt.plot(grid, res)
```

```python
hann_window = np.hanning(len(grid)) * res
plt.plot(grid, np.log10(np.abs(np.fft.fftshift(hann_window)))*10)
plt.plot(grid, f(grid))
```

```python
from scipy import signal

rect_window = res*signal.windows.boxcar(len(grid))
plt.plot(grid, np.log10(np.abs(np.fft.fftshift(rect_window)))*10)
plt.plot(grid, f(grid))

```
